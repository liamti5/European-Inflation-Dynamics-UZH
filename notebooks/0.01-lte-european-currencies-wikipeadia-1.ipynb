```python
import pandas as pd
```

```python
df = pd.read_csv("../data/external/european-currencies-wikipeadia.csv")
df
```

```python
df_duplicates = df[df.duplicated(subset=["Code", "Currency"])]
df_duplicates
```

The only countries that share currencies besides the Euro is Switzerland and Lichtenstein with the Franc (CHF). Since they are neighbouring countries, we will not differentiate between them when analysing Long/Lat.

```python
df_euro_countries = df[df["Code"] == "EUR"][["Country", "Currency", "Code"]]
df_euro_countries.to_csv("../data/interim/countries-using-EUR.csv", index=False)
df_euro_countries
```

```python
df_unique_currencies = df[["Country", "Currency", "Code"]].drop_duplicates(subset=["Code", "Currency"]).sort_values(by="Code")
df_unique_currencies
```
